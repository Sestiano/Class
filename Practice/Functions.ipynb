```python
# Sentiment analyzer with Polarity Detection

# Task: create a function that takes a sentence and analyze it by counting the number of positive
# and negative words based on predefined lists of positive and negative words

# Define two lists: one for positive word and one for negatigive ones. The function should return 
# a dictionary showing the count of positive and negative words in the sentence



positive_words = ["happy", "love", "great", "joy", "wonderful"]
negative_words = ["sad", "hate", "bad", "angry", "terrible"]

def sentiment_analyzer(sentence):
    words = sentence.lower().split(' ')
    print(words)
    counterp = 0
    countern = 0
    for word in words:
        if word in positive_words:
            counterp += 1
        if word in negative_words:
            countern += 1
    return {'positive': counterp, 'negative': countern}
    

print(sentiment_analyzer("I love programming but i Hate bugs"))


# Expected output:
# {'positive': 1, 'negative': 1}

```

```python
# Homeopathy function
# This function should dilute the solution untill the solution
# is diluted enuough to be considered a homeopathy solution.
# it should return the number of times the solution is diluted.

# lets say that the solution is diluted by 10 each time and
# it is considered a homeopathy solution if it is almost 1e-10 not under.
# so there is the fuction header and we need to implement the function.
# its easier to implement it with a recursive function.

def homeopathy(solution, dilution=10, limit=1e-10):
    if solution < limit:
        return 0
    solution /= dilution
    return homeopathy(solution, dilution, limit) + 1

print(homeopathy(1000))
```

```python
# Create a Python script that prints the Fibonacci series up to 10 using loops.
# The Fibonacci series is a sequence of numbers where each number is the sum of
# the two preceding ones, starting from 0 and 1.

def Fibonacci(n):
    first = 0
    second = 1
    print(f"{first}+{first}={first}")
    print(f"{first}+{second}={second}")
    while n > 0:
        third = first + second
        print(f"{first}+{second}={third}")
        first = second
        second = third
        n = n - 1

Fibonacci(10)
    
```

```python
import random
random.seed(0)


def create_matrix(size_n=10, alpha=0.4):
    matrix = {}
    for i in range(size_n):
        for j in range(size_n):
            if random.random() < alpha:
                matrix[(i, j)] = 1
            else:
                matrix[(i, j)] = 0
    return matrix

def print_values(matrix, size_n=10):
    for i in range(size_n):
        for j in range(size_n):
            print(matrix[(i, j)], end=' ')
        print('')


def print_matrix(matrix, size_n=10):
    for i in range(size_n):
        for j in range(size_n):
            if matrix[(i, j)] == 1:
                print('*', end=' ')
            else:
                print('-', end=' ')
        print('')




matrix = create_matrix(10, 0.4)
print_values(matrix, 10)
print_matrix(matrix, 10)
```

```python
def dead_or_alive(matrix, i, j, size_n=10):
    # edge cases:
    k_start = i-1 if i > 0 else 0
    k_end = i+1 if i < size_n-1 else size_n-1
    l_start = j-1 if j > 0 else 0
    l_end = j+1 if j < size_n-1 else size_n-1

    neighbors = 0
    for k in range(k_start, k_end+1):
        for l in range(l_start, l_end+1):
            if k == i and l == j:
                continue
            if matrix[(k, l)] == 1:
                neighbors += 1

    if neighbors < 2:
        return 0
    if neighbors > 3:
        return 1
    else:
        return matrix[(i, j)]


def play_game_of_life_step(matrix, size_n=10):
    new_matrix = {}
    for i in range(size_n):
        for j in range(size_n):
            new_matrix[(i, j)] = dead_or_alive(matrix, i, j, size_n)
    return new_matrix


print('initail state')
matrix = create_matrix(10, 0.3)
print_matrix(matrix, 10)

for gen_id in range(1, 11):
    print(f'Generation {gen_id}')
    matrix = play_game_of_life_step(matrix, 10)
    print_matrix(matrix, 10)
```
